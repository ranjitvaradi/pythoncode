```python
import matplotlib.pyplot as plt #Importing the required libraries 
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sales_data_with_discounts.csv") #importing the sales data from input file
nc=df.select_dtypes(include=['int','float']).columns.tolist() #Listing out numerical data
nc
```

```python
plt.figure(figsize=(10,3))#Figure Size
plt.boxplot(df['Volume'], vert=False) #plot chart for volume column
plt.title('Boxplot for volume') #Title
plt.xlabel('Volume Values')  #label x axis
```

```python
The central tendency of distribution is between 3 and 6.
The middle value of the data when arranged in ascending order (median) is 4.
The interquartile range lies between 3 and 6 and the whiskers extend between 1 to 10.
The outlayers are present between 10 and 30 units.
A longer whisker on one side of the box compared to the other side may indicate skewness in the distribution that is at right side.
```
