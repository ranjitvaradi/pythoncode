```python
import numpy as np  #Importing libraries
import pandas as pd
import random
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_data_with_discounts.csv")   #read from input file
nc=df.select_dtypes(exclude=['int','float']).columns.tolist()
nc1=df.select_dtypes(include=['int','float']).columns.tolist()
nc1
```

```python
gd = df.groupby('SKU')['Total Sales Value'].sum()
gd.plot(x='SKU', y='Total Sales Value', kind='bar', color='skyblue',figsize=(15,5))
```

```python
Inferences:

The categorical colum Stock per Unit is more on the unit M07 and less on F08.
As the order of SKU increases the units in each SKU increasing.
There is a huge increase or anomaly betwen L10 and M01 and between M06 and M07.
Unusually tall or short bars relative to other bars may indicate outliers that are F08 and M07.
```

```python
gd = df.groupby('BU')['Total Sales Value'].sum()
gd.plot(x='SKU', y='Total Sales Value', kind='bar', color='skyblue',figsize=(10,5))
```

```python
Inferences:

The categorical colum BU is more on Mobiles and less on FMCG.
As the order of BU increases the units in each BU increasing.
There is a huge increase or anomaly betwen Lifestyle and Mobiles.
Unusually tall or short bars relative to other bars may indicate outliers that is Mobiles.
```

```python
gd = df.groupby('Brand')['Total Sales Value'].sum()
gd.plot(x='BRAND', y='Total Sales Value', kind='bar', color='skyblue',figsize=(5,5))
```

```python
Inferences:

The Orange brand is the most and Gear is the least. 
There is a huge increase or anomaly between Jeera and Orange.
Unusually tall or short bars relative to other bars may indicate outliers that are Vedic and Orange.
```
