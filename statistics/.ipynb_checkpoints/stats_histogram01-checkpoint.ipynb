```python
NUMBER 02:
Data Visualization:
笳十tObjective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
笳十tHistograms:
	Plot histograms for each numerical column.
	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.
```

```python
import matplotlib.pyplot as plt #Importing required libraries and reading data from the input file
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df1=pd.read_csv("sales_data_with_discounts.csv")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4)) #Deviding the plot in to subplots for better analysis
axs[0].hist(df1["Volume"], bins=30, alpha=0.7,edgecolor='black') #at axis 0 histogram and axis 1 sns Kde bins are 30 , alpha is transperency of bars.
axs[0].set_title('Histogram of Sales Data for Volume') #Title
sns.histplot(df1["Volume"], bins=30, alpha=0.7, color='green',kde=True, ax=axs[1]) #for KDE chart
axs[0].set_xlabel('Volume') #Labels at x and y axis
axs[0].set_ylabel('count')
axs[1].set_title('Histogram of Sales Data for Volume With KDE')  #Title
plt.tight_layout()# This will adjust the space between charts so that they will not coinside.
plt.show()
```

```python
df1['Volume'].skew(),df1['Volume'].kurtosis()  #calculating skew ness and kurtosis
```

```python
Inferences:
The Volume distribution is skewed towards right side and the skewness is +ve and it is 2.73.
The central tendency of histogram is distributed at 5 and the maximum value is at 4.
The histogram is narrower and it  suggests lower variability or more concentrated data.The kurtosis value is positive and it is 10.25. 
As there isolated bars that indicates the outlayers
The histogram is unimodal as there is single peak at 4 means at 4 maximum value is recorded.
The most of the volume is in between the values 3 and 6.
```
