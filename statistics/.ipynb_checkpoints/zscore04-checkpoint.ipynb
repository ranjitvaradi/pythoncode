```python
import numpy as np
import pandas as pd
import random
from scipy import stats
from scipy.stats import zscore
```

```python
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
#Calculating zscore for each numerical columns 

zscore_volume = (df['Volume'] - df['Volume'].mean()) / df['Volume'].std()
zscore_Avgprice= (df['Avg Price'] - df['Avg Price'].mean()) / df['Avg Price'].std()
zscore_TotalSalesValue = (df['Total Sales Value'] - df['Total Sales Value'].mean()) / df['Total Sales Value'].std()
zscore_DiscountRate = (df['Discount Rate (%)'] - df['Discount Rate (%)'].mean()) / df['Discount Rate (%)'].std()
zscore_DiscountAmount = (df['Discount Amount'] - df['Discount Amount'].mean()) / df['Discount Amount'].std()
zscore_NetSalesValue = (df['Net Sales Value'] - df['Net Sales Value'].mean()) / df['Net Sales Value'].std()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(zscore_volume, bins=30, alpha=0.7, color='green',kde=True)
```

```python
Inferences:
Volume beyond 3 and -3 are treated as outlayers and hence there are 3 outlayers in the above graph.
Maximun volume is in between -1 and +1.
Data points with Z-scores close to 0 are closer to the mean and exhibit less deviation from normality, 
    while extreme Z-scores indicate more significant deviations in volume that are between 4 and 6.
```

```python
sns.histplot(zscore_Avgprice, bins=30, alpha=0.7, color='green',kde=True)
```

```python
Inferences:
Values beyond 3 and -3 are treated as outlayers and hence there no outlayers in the above graph.
Maximun volume is in between -0.5 and +0.5.
Data points with Z-scores close to 0 are closer to the mean and exhibit less deviation from normality, 
    while extreme Z-scores indicate more significant deviations that are between 2 and 3.
The initial values are having peak values among all.
```

```python
sns.histplot(zscore_TotalSalesValue, bins=30, alpha=0.7, color='green',kde=True)
```

```python
Inferences:
Total Sales Value beyond 3 and -3 are treated as outlayers and hence there are NO outlayers in the above graph.
Maximun Discount Rate% is in between -1 and +1.
Data points with Z-scores close to 0 are closer to the mean and exhibit less deviation from normality, 
    while extreme Z-scores indicate more significant deviations thaat are between 2 and 3.
The initial values are having peak values among all.
```

```python
Inferences:
Discount Rate% Beyond 3 and -3 are treated as outlayers and hence there are NO outlayers in the above graph.
Maximun Discount Rate% is in between -1 and +1.
Data points with Z-scores close to 0 are closer to the mean and exhibit less deviation from normality, 
    while extreme Z-scores indicate more significant deviations in Discount Rate% are between -0.5 and -2.5.
The skewness in on left side and -ve values.
```
