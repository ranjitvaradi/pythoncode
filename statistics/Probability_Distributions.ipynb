### Statistical Distributions
In statistics, the term "distribution" refers to the way in which values of a variable or a set of data are spread or distributed across different levels of measurement. Essentially, it describes the pattern of variation of a variable.

A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. Essentially, it is a function that assigns a numerical value to each outcome in the sample space of a random experiment. Random variables are used in statistics and probability theory to formally quantify and analyze the outcomes of random processes or experiments.

A probability distribution is a statistical function that describes all the possible values and likelihoods that a random variable can take within a given range. It provides a framework for predicting the probability of various outcomes in a random experiment. Probability distributions can be classified into two broad categories based on the nature of the random variable they describe: discrete probability distributions and continuous probability distributions.

#### 1.Discrete probability Distribution for tossing a coin twice (Binomial Experiment):

When you toss a coin twice, you're dealing with a discrete random variable because the number of possible outcomes is countable. Each toss of a fair coin has two possible outcomes: heads (H) or tails (T). Thus, when you toss the coin twice, the sample space, or the set of all possible outcomes, is:

S={HH,HT,TH,TT}

* The probability of getting no heads (TT), which is 1/4
* The probability of getting one heads (HT) or (TH) , which is 2/4
* The probability of getting two heads (HH), which is 1/4

The binomial experiment is characterized by having a fixed number of independent trials, each with only two possible outcomes (often termed as "success" and "failure"), and the probability of success being constant for each trial.

```python
import matplotlib.pyplot as plt

# Values of X
x_values = ['0H', '1H', '2H']
# Corresponding probabilities
probabilities = [0.25, 0.5, 0.25]

# Creating the bar plot
plt.figure(figsize=(6,3))
plt.bar(x_values, probabilities, color='skyblue')
plt.xlabel('Number of Heads (X)')
plt.ylabel('Probability')
plt.title('Probability Distribution of Heads in Two Coin Tosses')
plt.xticks(x_values)
plt.ylim(0, 1)

# Display the plot
plt.show()
```

#### 2. Discrete probability distribution for throwing dice twice
A single throw of a dice naturally has more than two outcomes since a standard die has six faces. The sample space for throwing a die once includes all six possible outcomes:

S1 = {1,2,3,4,5,6}
 
When throwing a die twice, you're looking at the set of all possible pairs of outcomes. The sample space for this scenario, which includes all combinations of the first and second throws, is represented as:

S2 = {(1,1),(1,2),(1,3),(1,4),(1,5),(1,6),(2,1),(2,2),…,(6,5),(6,6)}

This results in a total of 6×6=36 possible outcomes.


```python
import matplotlib.pyplot as plt

# Possible sums from 2 to 12
sum_values = list(range(2, 13))
# Probabilities for each sum
probabilities = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36]

# Creating the bar plot
plt.figure(figsize=(6,3))
plt.bar(sum_values, probabilities, color='green')
plt.xlabel('Sum of Outcomes (S)')
plt.ylabel('Probability')
plt.title('Probability Distribution for the Sum of Outcomes of Two Dice Throws')
plt.xticks(sum_values)
plt.ylim(0,max(probabilities) + 0.05)

# Display the plot
plt.show()
```

#### 3. Continuous probability distributions
A continuous probability distribution describes the probabilities of the possible values of a continuous random variable. A continuous random variable is one which can take an infinite number of possible values, often within a certain range. Unlike discrete random variables that have a countable number of possible outcomes, continuous random variables can assume any value within a continuous range. The concept of a continuous probability distribution is foundational in statistics and is used to model and analyze phenomena that vary continuously.

```python
# Dataset H
H = [10.5, 11.3, 9.6, 12.4, 18.6, 16.6, 13.9, 10.5]

# Count the number of values less than 13
less_than_13 = sum(1 for value in H if value < 13)

# Total number of values in H
total_values = len(H)

# Probability of finding H < 13
probability_less_than_13 = less_than_13 / total_values

probability_less_than_13
```

#### Continuos distribution curve for heights data
The histogram with the distribution curve for the 100 random height values generated between 140 cm and 185 cm. The histogram shows the distribution of these heights, while the smooth line (kernel density estimate, or KDE) overlays the distribution curve, providing a visual representation of the data's density across different height values. This combination offers both a discrete and a continuous perspective on how the heights are distributed

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 100 random height values between 140 cm and 185 cm
np.random.seed(42)  # For reproducibility
random_heights = np.random.uniform(low=140, high=185, size=100)
print(random_heights)
# Plotting the histogram and distribution curve on the same graph
plt.figure(figsize=(7, 4))
sns.histplot(random_heights, bins=15, kde=True, color='skyblue', alpha=0.6, stat="density")
plt.title('Histogram and Distribution Curve of 100 Random Height Values')
plt.xlabel('Height (cm)')
plt.ylabel('Density')

# Display the plot
plt.show()
```

#### A normal distribution (perfect / ideal distribution)
A normal distribution, also known as the Gaussian distribution, is a type of continuous probability distribution for a real-valued random variable. It is symmetric and has a bell-shaped curve, where the mean, median, and mode of the distribution are all equal. The normal distribution is characterized by two parameters namely mean and standard deviation: 

The mean determines the center of the distribution, and the standard deviation determines the spread of the distribution around the mean.

Key properties of the normal distribution include:

Symmetry: The distribution is symmetric about its mean, meaning the left half of the distribution is a mirror image of its right half.

Bell Shape: The highest point on the distribution curve corresponds to the mean, and the curve tapers off as it moves away from the mean, approaching but never touching the horizontal axis.

Defined by Mean and Standard Deviation: The entire shape of the distribution curve is determined by its mean and standard deviation.

Area Under the Curve: The total area under the curve equals 1. The area under the curve within a given range represents the probability of the variable falling within that range.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the normal distribution
mean = 162.5
std_dev = 10  # This is an assumption; adjust based on how spread out you want the distribution to be

# Generate 100 random height values
np.random.seed(42)  # For reproducibility
random_heights_normal = np.random.normal(loc=mean, scale=std_dev, size=100)

# Plotting the histogram and distribution curve
plt.figure(figsize=(8, 4))
sns.histplot(random_heights_normal, bins=15, kde=True, color='purple', alpha=0.6, stat="density")
plt.title('Histogram and Distribution Curve of 100 Random Height Values (Normal Distribution)')
plt.xlabel('Height (cm)')
plt.ylabel('Density')

# Display the plot
plt.show()
```

### Computation of probabilities from data distribution

In a continuous distribution, the y-axis represents the probability density function (PDF) which shows the probability per unit interval on the x-axis, not the frequency count. The height of the PDF curve at a given point indicates the relative likelihood of the random variable falling within a small range of values around that point.

So, while histograms can provide an approximation of the probability density, they are based on discrete bins and counts, whereas the PDF is a continuous function that describes the relative likelihood of different values for a continuous random variable.

The Cumulative Distribution Function (CDF) and the Probability Density Function (PDF) are both related to the distribution of a random variable in statistics, but they serve different purposes and convey different information about the distribution:

##### Probability Density Function
Applies to continuous random variables
The PDF describes the likelihood (density) of the random variable to take on a specific value. It is a function whose value at any given sample (or point) in the space of possible values can be interpreted as providing a relative likelihood that the value of the random variable would equal that sample.

###### Cumulative Distribution Function (CDF)
Applies to: Both discrete and continuous random variables.
What it shows: The CDF gives the probability that the random variable is less than or equal to a certain value. It is a function that maps values to their percentile rank in the distribution.

Interpretation: The CDF at a particular value x is the area under the PDF curve to the left of x (for continuous distributions) or the sum of probabilities up to x (for discrete distributions). It represents the probability that the random variable takes a value less than or equal to x.

Properties: The CDF starts at 0 and monotonically increases to 1 as you move from left to right across the range of possible values. For continuous distributions, it is the integral of the PDF from negative infinity to x.

##### Note
If you integrate the Probability Density Function (PDF) of a continuous random variable over an interval, you obtain the Cumulative Distribution Function (CDF) for the variable at the upper limit of that interval. 

```python
# Generate data of a normal distribution and plot it with pdf on y-axis, cdf on y-axis separately

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Define the range of x values
x = np.linspace(-4, 4, 1000)

# Define parameters for the normal distribution
mu = 0
sigma = 1

# Calculate the PDF and CDF values
pdf = stats.norm.pdf(x, mu, sigma)  # PDF values
cdf = stats.norm.cdf(x, mu, sigma)  # CDF values

# Plot PDF
plt.figure(figsize=(6, 3))
plt.plot(x, pdf, color='blue', label='PDF')
plt.title('Probability Density Function (PDF)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Plot CDF
plt.figure(figsize=(6, 3))
plt.plot(x, cdf, color='red', label='CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('X')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```
